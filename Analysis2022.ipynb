```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

### Data Structure
48部电影数据成分分析

```python
#  文件读取
films_info = pd.read_csv('.\data\info.csv')
films_info
```

#### Languages Structure

```python
languages = []
for i in films_info['language']:
    i = i.replace(' ','').split('/')
    languages += i
languages = Counter(languages)

plt.rcParams['font.sans-serif']=['SimHei'] # 正常显示中文
plt.figure(figsize = (8,8))  # 设置画布大小 
explode = (0.02, 0.03, 0.04, 0.1, 0.2, 0.3, 0.2, 0.1, 0.3, 0.1, 0.3, 0.2,0.1)  # 设置每一块分割出的间隙大小
plt.pie(x=languages.values(), labels=languages.keys(), explode = explode, autopct = '%1.2f%%',
        pctdistance = 0.8, labeldistance = 0.9)
plt.title("Languages")  # 设置标题
plt.axis('equal')
```

#### Types Structure

```python
types = []
for i in films_info['type']:
    i = i.replace(' ','').replace('[', '').replace(']','').replace("'","").split(',')
    types += i
types = Counter(types)

explode = (0.01,0.01,0.02,0.01,0.03,0.04,0.05,0.06,0.07,0.08,0.09,0.1,0.09,0.11,0.07,0.06,0.05,0.04,0.03,0.02,0.01)
plt.figure(figsize = (8,8))  # 设置画布大小 
plt.pie(x=types.values(), labels=types.keys(), explode=explode, autopct = '%1.2f%%',
        pctdistance = 0.8, labeldistance = 0.9)
plt.title("Types")  # 设置标题
plt.axis('equal')
```

#### Countries Structure

```python
countries=[]
for i in films_info['country']:
    i = i.replace(' ','').split('/')
    countries += i
countries = Counter(countries)

explode = (0.01, 0.01, 0.02, 0.01, 0.03, 0.04, 0.05, 0.06, 0.07, 0.08, 0.09, 0.1, 0.09)
plt.figure(figsize = (8,8))  # 设置画布大小 
plt.pie(x=countries.values(), labels=countries.keys(), explode=explode, autopct = '%1.2f%%',
        pctdistance = 0.8, labeldistance = 0.9)
plt.title("Countries")  # 设置标题
plt.axis('equal')
```

#### High Popularity
$$box-office>=200000$$

```python
def s_pie(idx):
    x = []
    for i in idx:
        if films_info["scores"][i] < 6:
            x.append("Low")
        elif films_info["scores"][i] >= 8:
            x.append("High")
        else:
            x.append("medium")
    x = Counter(x)
    
    plt.figure(figsize = (6,6))  # 设置画布大小 
    plt.pie(x=x.values(), labels=x.keys(), autopct = '%1.2f%%',
        textprops={'fontsize': 15}, pctdistance = 0.5, labeldistance = 0.8)
    plt.title("Scores")  # 设置标题
    plt.legend()
    plt.axis('equal')
```

```python
h_pop = []
for i, p in enumerate(films_info["box-office"]):
    if p >= 200000:
        h_pop.append(i)
h_pop
```

```python
s_pie(h_pop)
```

#### Low Scores
$$scores<6$$

```python
def p_pie(idx):
    x = []
    for i in idx:
        if films_info["box-office"][i] < 50000:
            x.append("Low")
        elif films_info["box-office"][i] >= 200000:
            x.append("High")
        else:
            x.append("medium")
    x = Counter(x)
    
    plt.figure(figsize = (6,6))  # 设置画布大小 
    plt.pie(x=x.values(), labels=x.keys(), autopct = '%1.2f%%',
        textprops={'fontsize': 15}, pctdistance = 0.5, labeldistance = 0.8)
    plt.legend()
    plt.title("Popularity")  # 设置标题
    plt.axis('equal')
```

```python
l_sco = []
for i, s in enumerate(films_info["scores"]):
    if s < 6:
        l_sco.append(i)
l_sco
```

```python
p_pie(l_sco)
```

#### High Scores
$$scores>=8$$

```python
h_sco = []
for i, s in enumerate(films_info["scores"]):
    if s >= 8:
        h_sco.append(i)
h_sco
```

```python
p_pie(h_sco)
```
