```python
import seaborn as sns
import pandas as pd
```

```python
films_info = pd.read_csv('.\data\info.csv')
f = []
for i in range(10):
    f.append(pd.read_csv('.\data\short-comments\\{}.csv'.format(i)))
```

```python
films_info
```

```python
#  电影票房和评分柱状图：2022，10部电影中仅有1部高分电影获得较高票房；其余无论评分高低，票房都比较低。
sns.barplot(x="scores", y="box-office", data=films_info, palette="YlGnBu")
```

```python
# 票房和评分散点图：2022，10部电影中仅有1部高分电影获得较高票房；其余无论评分高低，票房都比较低。
sns.scatterplot(x='scores', y='box-office', data=films_info, size="five")
```

```python
#  10部电影的评分统计直方图，电影总评分大致呈现正态分布样式：即高分和低分为少数，大多数评分属中游
sns.histplot(films_info['scores'], kde=True, bins=7)
```

```python
#  10部电影的票房统计直方图：可以看出2022斩获高票房的电影只是少数，大多数电影票房低迷
sns.histplot(films_info['box-office'], kde=True, bins=12)
```

```python
#  图表汇总
sns.jointplot(x='scores', y='box-office', data=films_info, kind='reg')
```

```python
sns.jointplot(x='scores', y='box-office', data=films_info, kind='kde', shade=True, cmap='YlGnBu')
```

```python
# correlation 关联性
films_info.corr()
```

```python
# heat map
# annotation 标注
sns.heatmap(films_info.corr(), annot=True, cmap='coolwarm')
```
